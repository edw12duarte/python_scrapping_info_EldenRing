# Scrapping url's para jefes elden ring en wiki

sacamos los nombres y url's de los jefes de elden ring de la pagina wiki 

```python
url = 'https://eldenring.wiki.fextralife.com/Bosses' #---> url de la pagina para hacer scraping
pagina = requests.get(url)                           #---> se crea un objeto tipo request
soup = BeautifulSoup(pagina.content,'html.parser')   #--> se cre una variable con todo el contenido de la pagina.  


#ESTOS 2 PASOS SON PARA BUSCAR UNAS ETIQUETAS ESPECFICAS EN EL document html
div_jefes_wiki = soup.find('div',{'class':'col-sm-6'}) #---> se busca la etiqueta padre especifica (div)
nombres_jefes_wiki =div_jefes_wiki.find_all('a')   #---> de la etiqueta anterior, se busca las etiquetas hijas que nos interese
 
```

```python

url_jefes_pagina = [i.get('href') for i in nombres_jefes_wiki][3:]
url_jefes_pagina
```

# Scrapping para info de jefes

```python
def traer_datos_url(url:str):
    url_pagina_wiki = 'https://eldenring.wiki.fextralife.com'

    # Traer todo el html de la pagina
    pagina = requests.get(url_pagina_wiki + url)                          
    soup_1 = BeautifulSoup(pagina.content,'html.parser')


    tabla_jefes = soup_1.find('table',{'class':'wiki_table'})
    info_jefes =tabla_jefes.findChildren('td')
    info_jefes_text = [i.text for i in info_jefes]
    
    return info_jefes_text
```

## Traer los datos en crudo

```python
todos_datos_jefes =[]

for jefe in url_jefes_necesito:
    todos_datos_jefes.append(traer_datos_url(jefe))
```

```python
todos_datos_jefes[0]
```

## Limpiar los datos en crudo

```python
def limpiar_datos_2(datos_jefes:list):
    for dato in range(len(datos_jefes)):
        datos_jefes[dato] = datos_jefes[dato].lower().replace('stronger vs','')
        datos_jefes[dato] = datos_jefes[dato].lower().replace('weak to','')
        datos_jefes[dato] = datos_jefes[dato].lower().replace('weaker to','')
        datos_jefes[dato] = datos_jefes[dato].lower().replace('strong vs','')
        datos_jefes[dato] = datos_jefes[dato].lower().replace('neutral','')
        datos_jefes[dato] = datos_jefes[dato].lower().replace('\n','<br>')

    return datos_jefes
```

```python
todos_datos_jefes_limpios=[]

for jefe in todos_datos_jefes:
    todos_datos_jefes_limpios.append(limpiar_datos_2(jefe))
```
